### Assignment 2

#### 1.Distance, JobRole and Attrition Breakdown

```python
df_home=df.loc[:,['DistanceFromHome','JobRole','Attrition']]
```

```python
#Jobrole and Distance by Total Attrition Counts
df_home.groupby(['JobRole','DistanceFromHome'])['Attrition'].count().sort_values(ascending=False)
```

```python
#Jobrole and Distance by Breakdown of Attrition Counts
df2=df_home.query('Attrition=="Yes"').groupby(['JobRole','DistanceFromHome'])['Attrition'].count().sort_values(ascending=False)
df2
```

```python
#Contingency Table For Jobroles with Breakdown on Attrition 
jroles=pd.crosstab(df.JobRole,df.Attrition).sort_values(by='Yes',axis=0,ascending=False)
jroles
```

```python
#Contingency Table For Jobroles and Distance showing Breakdown on Attrition 
df_distance=pd.crosstab([df.JobRole,df.DistanceFromHome],df.Attrition).sort_values(by='Yes',axis=0,ascending=False)
df_distance
```

```python
pd.crosstab(df.DistanceFromHome,df.Attrition).sort_index(axis=0,ascending=True)
```

#### 2. Education and Monthly Income Analysis

```python
def edu_numtostr(number):
    if number==1:
        return 'Below College'
    elif number==2:
        return 'College'
    elif number==3:
        return 'Bachelor'
    elif number==4:
        return 'Masters'
    elif number==5:
        return 'Doctorate'
```

```python
#.map(edu_numtostr)

df['Education']=df['Education'].apply(edu_numtostr)
```

```python
df_MonthIncome=df[['Education','MonthlyIncome','Attrition']]
```

```python
df_MonthIncome.groupby(['Education'])['MonthlyIncome'].mean().sort_values(ascending=True)
```

```python
df_MonthIncome.groupby(['Attrition'])['MonthlyIncome'].mean().sort_values(ascending=True)
```

```python
df_aveIncome=df_MonthIncome.groupby(['Education','Attrition'],as_index=False)['MonthlyIncome'].mean().sort_values(by='Education',ascending=True)
#df_aveIncome.rename({'MonthlyIncome':'Average Income'},inplace=True)
df_aveIncome
```

```python
df_aveIncome.values
```

```python
fig, ax = plt.subplots(figsize=(3, 5))
ax.set_ylabel( "Monthly Income" , size = 10 )
ax.set_title( "Monthly Average Income" , size = 12 )
sns.boxplot(y=df_aveIncome)
```
